```python
#Employee Class
class Employee:
    def __init__(self,name,age,position,salary):
        self.name=name
        self.age=age
        self.position=position
        self.salary=salary
    def give_raise(self,amount):
        self.salary+=amount
        print(f"Salary after appraisal is: {self.salary}")
    def promotion(self,new_position,raise_amount):
        self.position=new_position
        self.give_raise(raise_amount)
        print(f"Promoted to new position {new_position} and after promotion salary is {self.salary}")


E1=Employee("Faisal",26,"Junior Planner",4000)
E1.give_raise(2000)
E1.promotion("Planning Engineer",2000)


#Management System Class
class Management_System:
    def __init__(self):
        self.emp_list=[]

    def add_employee(self,employee):
        self.emp_list.append(employee)
        print(f"Employee {employee.name} is successfully added")

    def remove_employee(self,name):
        for emp in self.emp_list:
            if emp.name==name:
                self.emp_list.remove(emp)
                print(f"Employee {name} is successfully removed")
                return
        print(f"No employee in the list with {name}")

    def display_employees(self):
        if not self.emp_list:
            print(f"Employee List is Empty")
        else:
            for emp in self.emp_list:
                print(f"Name: {emp.name} , Age: {emp.age} , Position: {emp.position}, Salary: {emp.salary}")
    
    def annual_review(self):
        for emp in self.emp_list:
                if emp.age>40:
                    emp.give_rise(emp.salary*0.05)

                elif emp.position.lower()=="junior_developer":
                    emp.promotion("senior developer",emp.salary*0.10)

    def find(self,name):
        for emp in self.emp_list:
            if self.name==name:
                return emp
        return "Name not Found"


MS=Management_System()
MS.add_employee(E1)
MS.display_employees()
MS.remove_employee("Faisal")



#Main Programm Loop
while True:  
    print("\n--- Management System Menu ---")
    print("1. Add Employee")
    print("2. Remove Employee")
    print("3. Display Employees")
    print("4. Annual Review")
    print("5. Find Employee")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        position = input("Enter position: ")
        salary = float(input("Enter salary: "))
        emp = Employee(name, age, position, salary)  
        MS.add_employee(emp)  

    elif choice == "2":
        name = input("Enter the name of employee to remove: ")
        MS.remove_employee(name)

    elif choice == "3":
        MS.display_employees()

    elif choice == "4":
        MS.annual_review()
        print("Annual review completed.")

    elif choice == "5":
        name = input("Enter the name of employee to find: ")
        result = MS.find(name)
        if isinstance(result, Employee):
            print(f"Found: {result.name}, {result.position}, {result.salary}")
        else:
            print(result)

    elif choice == "6":
        print("Exiting program...")
        break  

    else:
        print("Invalid choice. Please enter a number from 1 to 6.")



```
